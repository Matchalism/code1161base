```python
HDI = pd.read_csv(filepath,
                           infer_datetime_format=True,
                           parse_dates=['2000'],
                           dayfirst=True)
HDI.head(2)
```

```python
increase = HDI[["1990","2015"]].groupby("1990").sum()

plt.xkcd()
plt.plot(increase, "x-")
plt.title("Index", fontsize=18)
plt.xlabel('Human Development Index increase times 1990-2015', fontsize=26)
plt.ylabel('Times', fontsize=26)
plt.grid(True)
plt.show()
```

```python
capped_1990_data = HDI["1990"][HDI["1990"] < 40]

plt.hist(capped_1990_data, bins=10, facecolor='blue', alpha=0.2)#Old
plt.hist(capped_1990_data, bins=50, facecolor='green', alpha=1)#New
plt.show()
```

```python
number_of_bins = 15
lower_bound = 0
upper_bound = 1

plt.hist(HDI["1990"], bins=number_of_bins, range=(lower_bound, upper_bound))
plt.xlabel('Index', fontsize=20)
plt.ylabel('Frequency', fontsize=20)
plt.grid(True)
plt.show()
```
